# Assignment 4 - Gradient Descent Algorithm

```python
# import required python libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# function to calculate value of y at x
def y(x):
    return (x+3)**2;
```

```python
# function to claculate derivative of y at x
def y_derivative(x):
    return 2*(x+3);
```

```python
# generate datapoints for plotting graph of y(x) 
x_values = []
y_values = []

for i in range(-10, 11):
    x_values.append(i)
    y_values.append(y(i))
    
print(x_values)
print(y_values)
```

```python
# plot graph of y(x) and starting point
start_x = 2
start_y = y(start_x)

sns.lineplot(x=x_values, y=y_values)
plt.xticks(x_values)
plt.plot(start_x, start_y, color="red", marker="o")
plt.show()
```

```python
# gradient descent algorithm to find local minima
minima_x = start_x
minima_y = start_y
iterations = 1000
learning_rate = 0.005

for i in range(iterations):
    minima_x -= learning_rate*y_derivative(minima_x)
    minima_y = y(minima_x)        
    print(minima_x, minima_y)
```

```python
# plot graph of y(x) and local minima
sns.lineplot(x=x_values, y=y_values)
plt.xticks(x_values)
plt.plot(minima_x, minima_y, color="red", marker="o")
plt.show()
```
